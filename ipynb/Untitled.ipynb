```python
def make_final_air_pollution(df, df_new, year, season, pollution, region):
    # 연도와 지역 기준으로 필터링
    df_filtered = df[(df['연도'] == year) & (df['지역'] == region)]
    
    # 오염종류 기준으로 필터링
    temp = df_filtered[df_filtered['오염종류'] == pollution]
    
    # 분기별 평균 및 분산 계산
    if season == 1:
        temp = temp.loc[:, ['1월', '2월', '3월']]
    elif season == 2:
        temp = temp.loc[:, ['4월', '5월', '6월']]
    elif season == 3:
        temp = temp.loc[:, ['7월', '8월', '9월']]
    elif season == 4:
        temp = temp.loc[:, ['10월', '11월', '12월']]
    
    avg = temp.mean().mean()
    var = temp.var().mean()
    if (pollution == '이산화황'):
        new_row = {
            '연도': year,
            '분기': season,
            '지역': region,
            pollution + '_평균': avg,
            pollution + '_분산': var
        }
        pollutants = ['이산화질소', '오존', '일산화탄소', '미세먼지', '초미세먼지']
        for pol in pollutants:
            if pol != pollution:
                new_row[pol + '_평균'] = 0
                new_row[pol + '_분산'] = 0
        df_new = pd.concat([df_new, pd.DataFrame([new_row])], ignore_index=True)
    else:
        # 기존 행을 업데이트
        index = df_new[(df_new['연도'] == year) & (df_new['분기'] == season) & (df_new['지역'] == region)].index
        if not index.empty:
            idx = index[0]
            df_new.at[idx, pollution + '_평균'] = avg
            df_new.at[idx, pollution + '_분산'] = var
    
    
    return df_new
```

```python
def make_final_air_pollution(df, df_new, year, season, pollution, region):
    # 연도와 지역 기준으로 필터링
    df_filtered = df[(df['연도'] == year) & (df['지역'] == region)]
    
    # 오염종류 기준으로 필터링
    temp = df_filtered[df_filtered['오염종류'] == pollution]
    
    # 분기별 값 선택
    if season == 1:
        temp = temp.loc[:, ['1월', '2월', '3월']]
    elif season == 2:
        temp = temp.loc[:, ['4월', '5월', '6월']]
    elif season == 3:
        temp = temp.loc[:, ['7월', '8월', '9월']]
    elif season == 4:
        temp = temp.loc[:, ['10월', '11월', '12월']]
        
    # 모든 값을 한 시리즈로 변환하여 평균 및 분산 계산
    all_values = temp.values.flatten()
    all_values = all_values[~np.isnan(all_values)]  # NaN 값 제거
    avg = all_values.mean()
    var = all_values.var()
    
    if pollution == '이산화황':
        new_row = {
            '연도': year,
            '분기': season,
            '지역': region,
            pollution + '_평균': avg,
            pollution + '_분산': var
        }
        pollutants = ['이산화질소', '오존', '일산화탄소', '미세먼지', '초미세먼지']
        for pol in pollutants:
            new_row[pol + '_평균'] = 0
            new_row[pol + '_분산'] = 0
        df_new = pd.concat([df_new, pd.DataFrame([new_row])], ignore_index=True)
    else:
        # 기존 행을 업데이트
        index = df_new[(df_new['연도'] == year) & (df_new['분기'] == season) & (df_new['지역'] == region)].index
        if not index.empty:
            idx = index[0]
            df_new.at[idx, pollution + '_평균'] = avg
            df_new.at[idx, pollution + '_분산'] = var
    
    return df_new
```
