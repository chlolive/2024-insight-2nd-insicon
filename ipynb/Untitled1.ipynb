```python
import pandas as pd
import numpy as np
```

```python
hum = pd.read_csv("humidity.csv")
hum
```

```python
hum.drop('중앙값',axis=1,inplace=True)
hum = hum.rename(columns={'평균':'습도_평균', '분산':'습도_분산'})
hum
```

```python
tem_rain = pd.read_excel("./꿀벌_기온_강수량.xlsx")
tem_rain
```

```python
tem_rain.columns
```

```python
tem_rain.drop(['최고기온', '최저기온', '기온_중앙값','최고강수량', '최저강수량', '강수량_중앙값'],axis=1, inplace=True)
tem_rain
```

```python
tem_rain = tem_rain.rename(columns={'평균기온':'기온_평균', '평균강수량':'강수량_평균'})
tem_rain
```

```python
air = pd.read_csv("air_pollution_final.csv")
air
```

```python
air_new = pd.read_excel("대기오염물질_2023.xlsx")
air_new = air_new.rename(columns={'도시':'지역'})
air_new
```

```python
air_total = pd.concat([air_new, air], ignore_index=True)
air_total
```

```python
df_merge = pd.merge(hum, tem_rain, on=["연도","분기","지역"], how="inner")
df_merge2 = pd.merge(df_merge, air_total, on=["연도","분기","지역"], how="inner")
df_merge2
```

```python
df_merge2.columns
```

```python
# '_분산'이 포함된 모든 열 이름을 가져옴
variance_columns = [col for col in df_merge2.columns if '_분산' in col]

# 각 '_분산' 열에 대해 루트를 씌워서 '_표준편차' 열을 생성
for col in variance_columns:
    std_col = col.replace('_분산', '_표준편차')
    df_merge2[std_col] = np.sqrt(df_merge2[col])

df_merge2
```

```python
# '_분산'이 포함된 모든 열 이름을 가져옴
variance_columns = [col for col in df_merge2.columns if '_분산' in col]
df_merge2.drop(variance_columns, axis=1,inplace=True)
```

```python
df_merge2
```

```python
disease = disease[['연도','분기','지역','낭충봉아부패병']]
disease
```

```python
disease['낭충봉아부패병'].value_counts()
```

```python
df_merge3 = pd.merge(df_merge2, disease, on=["연도","분기","지역"], how="left")
df_merge3
```

```python
disease_new = pd.read_excel("꿀벌_낭충봉아발생여부.xlsx")
disease_new
```

```python
disease_new.loc[:,'낭충봉아부패병'] = disease_new['낭충봉아부패병'].apply(lambda x: 1 if x >= 1 else 0)
disease_new
```
