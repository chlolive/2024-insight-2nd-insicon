```python
air = pd.read_csv("./air_pollution_final.csv")
air
```

```python
air[air['지역']=='경기']
```

```python
def total_region_df(region):
    df1 = hum[hum['지역']==region]
    df1 = df1.rename(columns={'평균':'습도_평균', '중앙값':'습도_중앙값', '분산':'습도_분산'})
    
    df2 = tem_rain[tem_rain['지역']==region]
    df2 = df2.rename(columns={'평균기온':'기온_평균', '최고기온':'기온_최고', '최저기온':'기온_최저', 
                                '평균강수량':'강수량_평균', '최고강수량':'강수량_평균', '최저강수량':'강수량_최저'})
    df2 = df2.drop(columns='지역')
    
    df3 = disease[['연도','분기',region]]
    df3 = df3.rename(columns={region:'질병건수'})
    df3['연도'] = df3['연도'].str.replace('년', '').astype(int)

    df4 = air[air['지역']==region]
    df4 = df4.drop('지역',axis=1)
    
    df_merge = pd.merge(df1, df2, on=["연도","분기"], how="inner")
    df_total = pd.merge(df_merge, df3, on=["연도","분기"], how="left")
    df_total2 = pd.merge(df_total, df4, on=["연도","분기"], how="left")

    df_total2 = df_total2[df_total2['연도'] != 2023]
    return df_total2
```

```python
gyeonggi = total_region_df('충남')
gyeonggi
```

```python
regions = ['서울', '경기', '강원', '충북', '충남', '전북', '전남', '경북', '경남', '제주']
for region in regions:
    df_final = total_region_df(region)
    file_path_title = "./" + region + ".csv"
    df_final.to_csv(file_path_title)
```
